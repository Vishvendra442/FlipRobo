# Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file. 

```python
#Making empty list of all the data we need
titles=[]
imdb_ratings=[]
years=[]
```

```python
#Requesting URL
num=('0','51')
```

```python
for p in num:
    page=requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=' + str(p) + '&ref_=adv_nxt')
    soup = BeautifulSoup(page.text, 'html.parser')
    
    #selecting whole container which has all info
    movie_block = soup.find_all('div', class_='lister-item mode-advanced')
    
    for i in movie_block:
        # Scraping the movie's name
        name = i.h3.a.text
        titles.append(name)
        
        # Scraping the movie's year
        year = i.h3.find('span', class_='lister-item-year').text
        years.append(year)
        
        # Scraping the rating
        imdb = float(i.strong.text)
        imdb_ratings.append(imdb)
        
```

```python
#making dataframe
movies = pd.DataFrame({'movie':titles,'year':years,'imdb_rating':imdb_ratings})
movies
```

```python
#saving dataframe to csv
movies.to_csv('movies_list.csv')
```

# Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file. 

```python
page=requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
page
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
#Making empty list of all the data we need
titles=[]
imdb_ratings=[]
years=[]
```

```python
#selecting whole container which has all info
movie_block = soup.find_all('td', class_="titleColumn")
for i in movie_block:
    # Scraping the movie's name
    titles.append((i.a.text).replace('\n',''))
        
    # Scraping the movie's year
    years.append((i.span.text).replace('\n',''))
```

```python
# Scraping the rating
rating = soup.find_all('td',class_="ratingColumn imdbRating")
for i in rating:
    imdb_ratings.append((i.text).replace('\n',''))
```

```python
#making dataframe
movies = pd.DataFrame({'movie':titles,'year':years,'imdb_rating':imdb_ratings}).head(100)
movies
```

```python
#saving dataframe to csv
movies.to_csv('movies_indian_list.csv')
```

# Write a python program to scrap book name, author name, genre and book review of any 5 books from www.bookpage.com


```python
page=requests.get('https://bookpage.com/reviews')
page
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
#Making empty list of all the data we need
name=[]
author=[]
gener=[]
review=[]
```

```python
#selecting whole container which has all info
book_block = soup.find_all('div', class_='flex-article-content')
for i in book_block:
    # Scraping the book name
    t1 = i.h4.a.text
    name.append(t1)
        
    # Scraping the author
    au = i.find('p', class_='sans bold').text.replace('\n','')
    author.append(au)
    
    # Scraping the gener
    ge = i.p.text
    gener.append(ge)
    
    # Scraping the review
    re = i.find('p', class_='excerpt').text.replace('\n','')
    review.append(re)
```

```python
#making dataframe
books = pd.DataFrame({'Book':name,'Author':author,'Gener':gener, 'Review':review}).head(5)
books
```

# Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:


i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating. 

```python
def ranking(url):
    page= requests.get(url)


    soup = BeautifulSoup(page.text, 'lxml')

    table = soup.find('table',class_='table')


    name=[]
    team=[]
    rating=[]
    points=[]
    for row in table.find_all('tr'):
        link = row.find_all('td')
        if len(link) == 5:
            name.append(link[1].text.replace('\n',''))
            team.append(link[2].text.replace('\n',''))
            rating.append(link[3].text.replace('\n',''))
            points.append(link[4].text.replace('\n',''))
    names=pd.DataFrame(name)
    teams=pd.DataFrame(team)
    ratings=pd.DataFrame(rating)
    points=pd.DataFrame(points)
    
    
    df=pd.concat([names,teams,ratings,points],axis=1)            
    df=df.head(10)
    return df


```

```python
top10teams=ranking("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
```

```python
top10teams.columns=['Team', 'Matches', 'Point','Rating']
top10teams
```

# Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:

```python
def ranking1(url):
    page= requests.get(url)


    soup = BeautifulSoup(page.text, 'lxml')

    table = soup.find('table',class_='table')


    name=[]
    team=[]
    rating=[]
    points=[]
    for row in table.find_all('tr'):
        link = row.find_all('td')
        if len(link) == 5:
            name.append(link[1].text.replace('\n',''))
            team.append(link[2].text.replace('\n',''))
            rating.append(link[3].text.replace('\n',''))
            points.append(link[4].text.replace('\n',''))
    names=pd.DataFrame(name)
    teams=pd.DataFrame(team)
    ratings=pd.DataFrame(rating)
    points=pd.DataFrame(points)
    
    
    df=pd.concat([names,teams,ratings,points],axis=1, join='inner')            
    df=df.head(10)
    return df
```

# Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
page=requests.get('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YB_5b-gzbIU')
page
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
#Making empty list of all the data we need
period=[]
short=[]
temp=[]
desc=[]
```

```python
#selecting whole container which has all info
block = soup.find(id ="seven-day-forecast-container")
stone =block.find_all('div', class_='tombstone-container')
for i in stone:
    # Scraping the Peroid
    t1 = i.find('p', class_='short-desc').text
    period.append(t1)
        
    # Scraping the Short Description
    t2 = i.find('p', class_='short-desc').text
    short.append(t2)
    
    # Scraping the Temperature
    t3 = i.find('p', class_='temp').text.split()[-2]
    temp.append(t3)
    
    # Scraping the Description
    description = soup.find('div', class_='col-sm-10 forecast-text').text
    desc.append(description)
```

```python
#making dataframe
forecast = pd.DataFrame({'Period':period,'Short Description':short,'Temperature in F':temp, 'Description':desc}).head(7)
forecast
```

# Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
#here we are taking data of first two page but by typing range of page we can extract data of as many page we want
num= np.arange(1, 2)
```

```python
#Making empty list of all the data we need
name=[]
prices=[]
ratings=[]
url_img=[]
```

```python
for p in num:
    page=requests.get('https://www.amazon.in/s?i=electronics&bbn=1389432031&rh=n%3A976419031%2Cn%3A1389401031%2Cn%3A1389432031%2Cn%3A1805560031%2Cp_36%3A-2000000&dc&page='+str(p)+'&qid=1612724651&rnid=1389432031&ref=sr_pg_'+str(p))
    soup = BeautifulSoup(page.text, 'html.parser')
    print(page)
    
    #selecting whole container which has all info
    result=soup.find_all('div', {'data-component-type':'s-search-result'} )
    
    for it in result:
        
        #extracting name
        atag=it.h2.a.span.text
        name.append(atag)
        
        #extracting price
        try:
            parentprice = it.find('span', class_='a-price')
            price = parentprice.find('span', class_='a-offscreen').text
            prices.append(price)
        except AttributeError:
            price=''
            
        #extracting rating
        try:
            rating = it.i.text
            ratings.append(rating)
        except AttributeError:
            rating=''
            
        #extracting image url
        image = it.find('div', class_='a-section')
        img_url = image.img.get('src')
        url_img.append(img_url)
```

```python
#making dataframe
smartphone = pd.DataFrame({'Product':name,'Price':prices,'Rating':ratings, 'Image URL':url_img})
smartphone
```
